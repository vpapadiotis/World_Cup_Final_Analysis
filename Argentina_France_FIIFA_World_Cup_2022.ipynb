```python
f_xg = [0]
a_xg = [0]
f_min = [0]
a_min = [0]
```

```python
argentina = xg_data['team'].iloc[0]
france = xg_data['team'].iloc[-2]

# Iterate over DataFrame rows
for index, row in xg_data.iterrows():
  try:
    if row['team'] == france:
      f_xg.append(row['xG'])
      f_min.append(row['minute'])
    elif row['team'] == argentina:
      a_xg.append(row['xG'])
      a_min.append(row['minute'])
    else:
      print("Unknown team found in row:", index)
      print(row)
  except Exception as e:
    print("Error processing row:", index)
    print(e)

def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

f_cumulative = nums_cumulative_sum(f_xg)
a_cumulative = nums_cumulative_sum(a_xg)

f_last = round(f_cumulative[-1],2)
a_last = round(a_cumulative[-1],2)
```

```python
print("Cumulative xg for France:", f_cumulative)
print("Cumulative xg for Argentina:", a_cumulative)
print("Last cumulative xg for France:", f_last)
print("Last cumulative xg for Argentina:", a_last)
```

```python
# Draw the expected goals flow graph
fig, ax = plt.subplots(figsize=(10, 5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

#Set up base layer
mpl.rcParams['xtick.color'] = 'white'
mpl.rcParams['ytick.color'] = 'white'

ax.grid(ls='dotted', lw=.5, color='lightgrey', axis='y', zorder=1)
spines = ['top', 'bottom', 'left', 'right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)

plt.xticks([0, 15, 30, 45, 60, 75, 90, 105, 120])
plt.xlabel('Minute', fontname='Andale Mono', color='white', fontsize=16)
plt.ylabel('Expected Goals', fontname='Andale Mono', color='white', fontsize=16)

# Plot the step graphs
ax.step(x=f_min, y=f_cumulative, color='blue', label='France', linewidth=2, where='post')
ax.step(x=a_min, y=a_cumulative, color="#ADD8E6", label='Argentina', linewidth=2, where='post')

plt.legend(loc='upper left', fontsize=12)
plt.title('FIFA World Cup Final 2022 - Expected Goals Flow Chart', fontsize=18, color='white')
plt.show()
```

```python
def create_shot_df(df, argentina, france):
    shots_argentina = shots[shots.team == argentina]
    shots_france = shots[shots.team == france]
    return (shots_argentina, shots_france)

shots_argentina, shots_france = create_shot_df(shots, argentina, france)
```

```python
def create_statistics(shots):
    outcomes = shots.shot_outcome.value_counts()
    all_shots = outcomes.sum()
    goals = outcomes.Goal if hasattr(outcomes, 'Goal') else 0
    saves = outcomes.Saved if hasattr(outcomes, 'Saved') else 0
    post = outcomes.Post if hasattr(outcomes, 'Post') else 0
    on_target = saves + goals
    off_target = all_shots - on_target
    blocked = outcomes.Blocked if hasattr(outcomes, 'Blocked') else 0
    xg = round(shots.shot_statsbomb_xg.sum(), 2)

    team_stats = {'all_shots': all_shots, 'on_target': on_target, 'off_target': off_target, 'goals': goals, 'saves': saves, 'blocked': blocked, 'xg': xg}
    return team_stats

argentina_stats = create_statistics(shots_argentina)
france_stats = create_statistics(shots_france)
```

```python
def plot_shots(ax, shots, color, pitch_length_x, argentina=True):
    marker_map = {'Goal': "*", 'Saved': "o", 'Off T': "x", 'Blocked': "s", 'Wayward':"x", 'Post': "x"}

    locations = shots.location.values
    outcomes = shots.shot_outcome
    shot_xgs = shots.shot_statsbomb_xg.values

    for loc, outcome, xg in zip(locations, outcomes, shot_xgs):
        x = loc[0]
        y= loc[1]
        size = xg * 1000

        if argentina:
            x_cor = pitch_length_x - x + random.randint(-9, 9)/10
            y_cor = y + random.randint(-9, 9)/10
            pitch.scatter(x_cor, y_cor, s=size, color=color, alpha=.8, marker=marker_map[outcome], ax=ax)
        else:
            x_cor = x + random.randint(-9, 9)/10
            y_cor = y + random.randint(-9, 9)/10
            pitch.scatter(x_cor, y_cor, s=size, color=color, alpha=.8, marker=marker_map[outcome], ax=ax)
```

```python
def add_xgoal_desc(line_color, ax):
    ax.text(10, 75, 'xG', va='center', ha='center', color=line_color, fontsize=12, fontweight='bold')
    ax.arrow(4, 78, 20, 0, head_width=2, head_length=1, ec=line_color, fc=line_color, width=0.2)

    ax.add_patch(plt.Circle((17, 75), 0.2, color=line_color))
    ax.add_patch(plt.Circle((19, 75), 0.4, color=line_color))
    ax.add_patch(plt.Circle((21.5, 75), 0.8, color=line_color))
```

```python
def add_title(argentina, france, argentina_color, france_color, line_color, subtitle, ax):
    title = f"<{argentina}> <vs> <{france}>"
    title_font = 15
    subtitle_font = 13
    highlight_textprops =[
        {"color": argentina_color, "fontsize":title_font, "fontweight":'bold'},
        {"color": line_color, "fontsize":title_font,  "fontweight":'bold'},
        {"color": france_color, "fontsize":title_font,  "fontweight":'bold'}
    ]

    HighlightText(x=0.5, y=0.7, va='center', ha='center',
                s=title,
                highlight_textprops=highlight_textprops,
                ax=ax)

    ax.text(0.5, 0.0, subtitle, color=line_color, va='center', ha='center', fontsize=subtitle_font)
```

```python
def plot_legend(line_color, pitch_color):
    goal = Line2D([0], [0], marker='*', markersize=np.sqrt(30), color=line_color, linestyle='None')
    saved = Line2D([0], [0], marker='o', markersize=np.sqrt(30), color=line_color, linestyle='None')
    off_t = Line2D([0], [0], marker='x', markersize=np.sqrt(30), color=line_color, linestyle='None')
    blocked = Line2D([0], [0], marker='s', markersize=np.sqrt(30), color=line_color, linestyle='None')

    plt.legend([goal, saved, off_t, blocked], ['Goal', 'Saved', 'Off target','Blocked'], loc="lower right",
            markerscale=1.5, scatterpoints=1, fontsize=10, labelcolor=line_color, facecolor = pitch_color, edgecolor = line_color)
```

```python
def plot_stats(argentina_stats, france_stats, line_color, pitch_color, ax):
    stat_font_size = 12
    ax.text(x=60, y=5, s='{}    Goals    {}'.format(argentina_stats['goals'], france_stats['goals']), size=stat_font_size,
            color=line_color, backgroundcolor=pitch_color, ha='center')
    ax.text(x=60, y=10, s='{}    xGoal    {}'.format(argentina_stats['xg'], france_stats['xg']), size=stat_font_size,
            color=line_color, backgroundcolor=pitch_color, ha='center')
    ax.text(x=60, y=15, s='{}    Shots    {}'.format(argentina_stats['all_shots'], france_stats['all_shots']), size=stat_font_size,
            color=line_color,backgroundcolor=pitch_color, ha='center')
    ax.text(x=60, y=20, s='{}    On target    {}'.format(argentina_stats['on_target'], france_stats['on_target']), size=stat_font_size,
            color=line_color, backgroundcolor=pitch_color, ha='center')
    ax.text(x=60, y=25, s='{}    Off target    {}'.format(argentina_stats['off_target'], france_stats['off_target']), size=stat_font_size,
            color=line_color, backgroundcolor=pitch_color, ha='center')
    ax.text(x=60, y=30, s='{}    Saves    {}'.format(argentina_stats['saves'], france_stats['saves']), size=stat_font_size,
            color=line_color, backgroundcolor=pitch_color, ha='center')
    ax.text(x=60, y=35, s='{}    Blocked    {}'.format(argentina_stats['blocked'], france_stats['blocked']), size=stat_font_size,
            color=line_color, backgroundcolor=pitch_color, ha='center')
```

```python
shots_argentina, shots_france = create_shot_df(shots, argentina, france)
argentina_stats = create_statistics(shots_argentina)
france_stats = create_statistics(shots_france)

pitch = Pitch(pitch_type='statsbomb', pitch_color=pitch_color, line_color=line_color)
fig, axs = pitch.grid(figheight=6, title_height=0.05, endnote_space=0, axis=False,
                        title_space=0, grid_height=0.82, endnote_height=0.01)

fig.set_facecolor(pitch_color)

plot_shots(axs['pitch'], shots_argentina, argentina_color, pitch_length_x, argentina=True)
plot_shots(axs['pitch'], shots_france, france_color, pitch_length_x, argentina=False)

add_xgoal_desc(line_color, axs['pitch'])

subtitle = "2022 - Word Cup Final - Shots Map"
add_title(argentina,france, argentina_color, france_color, line_color, subtitle, axs['title'])

plot_legend(line_color, pitch_color)

plot_stats(argentina_stats, france_stats, line_color, pitch_color, axs['pitch'])
```
